```python
import pandas as pd
```

```python
dataset=pd.read_csv('CKD.csv')
dataset
```

```python
dataset=pd.get_dummies(dataset)
dataset=dataset.drop(['age','cad_no','pcc_notpresent','ba_notpresent','htn_no','dm_no','appet_poor','pe_poor','ane_no','classification_no'],axis=1)
dataset
```

```python
dataset['classification_yes'].value_counts()
```

```python
dataset.columns
```

```python
independent= dataset[['bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv', 'wc',
       'rc', 'sg_a', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_abnormal',
       'rbc_normal', 'pc_abnormal', 'pc_normal', 'pcc_present', 'ba_present',
       'htn_yes', 'dm_yes', 'cad_yes', 'appet_yes', 'pe_yes', 'ane_yes']]
independent
```

```python
dependent= dataset[['classification_yes']]
dependent
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=1/3,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test =  sc.fit_transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.model_selection import GridSearchCV
grid_parameters = {'solver':['newton-cg','lbfgs','liblinear','saga'],'penalty':['12']}
grid = GridSearchCV(LogisticRegression(),grid_parameters,refit=True,verbose=3,n_jobs=-1,scoring='f1_weighted')
grid.fit(x_train,y_train)
```

```python
from sklearn.svm import SVC
```

```python
from sklearn.model_selection import GridSearchCV
grid_parameters = {'kernel': ['linear','rbf','poly','sigmoid'],'gamma':['auto','scale'],'C':[10,100,1000,2000,3000]}
grid = GridSearchCV(SVC(probability=True),grid_parameters,refit=True,verbose=3,cv=5,n_jobs=-1,scoring='f1_weighted')
grid.fit(x_train,y_train)
```

```python
result = grid.cv_results_
grid_prediction = grid.predict(x_test)
```
