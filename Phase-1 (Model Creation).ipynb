```python
import pandas as pd
```

```python
dataset=pd.read_csv('CKD.csv')
dataset
```

```python
dataset=pd.get_dummies(dataset)
dataset=dataset.drop(['age','cad_no','pcc_notpresent','ba_notpresent','htn_no','dm_no','appet_poor','pe_poor','ane_no','classification_no'],axis=1)
dataset
```

```python
dataset['classification_yes'].value_counts()
```

```python
dataset.columns
```

```python
independent= dataset[['bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv', 'wc',
       'rc', 'sg_a', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_abnormal',
       'rbc_normal', 'pc_abnormal', 'pc_normal', 'pcc_present', 'ba_present',
       'htn_yes', 'dm_yes', 'cad_yes', 'appet_yes', 'pe_yes', 'ane_yes']]
independent
```

```python
dependent= dataset[['classification_yes']]
dependent
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=1/3,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test =  sc.fit_transform(x_test)
```

```python
# Random Forest
from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier(n_estimators=10,criterion='entropy',random_state=0)
classifier = classifier.fit(x_train,y_train)
```

```python
print(x_train)
```

```python
print(x_test)
```

```python
y_predict=classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test,y_predict)
confusion_matrix
```

```python
from sklearn.metrics import classification_report
classification_report = classification_report(y_test,y_predict)
print(classification_report)
```

```python
# Decison Tree
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion='entropy',random_state=0)
classifier = classifier.fit(x_train,y_train)
```

```python
y_predict=classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test,y_predict)
confusion_matrix
```

```python
from sklearn.metrics import classification_report
classification_report = classification_report(y_test,y_predict)
print(classification_report)
```

```python
# Support Vector Machine
from sklearn.svm import SVC
classifier = SVC(kernel='rbf',random_state=0)
classifier = classifier.fit(x_train,y_train)
```

```python
y_predict=classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test,y_predict)
confusion_matrix
```

```python
from sklearn.metrics import classification_report
classification_report = classification_report(y_test,y_predict)
print(classification_report)
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state=0)
classifier.fit(x_train,y_train)
```

```python
y_predict = classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test,y_predict)
confusion_matrix
```

```python
from sklearn.metrics import classification_report
classification_report = classification_report(y_test,y_predict)
print(classification_report)
```

```python
# K-Nearest Neighbour
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 7, metric = 'minkowski',p = 2)
classifier = classifier.fit(x_train,y_train)
```

```python
y_predict=classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test,y_predict)
confusion_matrix
```

```python
from sklearn.metrics import classification_report
classification_report = classification_report(y_test,y_predict)
print(classification_report)
```
